```python
import cv2
import numpy as np
import onnxruntime as ort
import matplotlib.pyplot as plt
%matplotlib inline

from utils.process import preprocessing, postprocessing

VIDEO_PATH = "data/videos/video.mp4"
ONNX_MODEL_PATH = "models/pfd.onnx"
CLASS_NAMES = ["BG", "Forklift", "Palete"]
IMAGE_SHAPE = (640, 640)
```

```python
session = ort.InferenceSession(ONNX_MODEL_PATH)
input_name = session.get_inputs()[0].name
output_name = session.get_outputs()[0].name
```

```python
cap = cv2.VideoCapture(VIDEO_PATH)

while True:
    ret, image = cap.read()

    tensor_input, ratios = preprocessing(image, IMAGE_SHAPE)

    predictions = session.run([output_name], {input_name: tensor_input})

    outputs = postprocessing(predictions[0], ratios)

    if len(outputs) >= 1:
        detections = outputs[0]

        if detections.ndim == 1:
            detections = detections.reshape(1, -1)

        for det in detections:
            x1, y1, x2, y2 = map(int, det[:4])
            class_id = int(det[8])
            score = det[4] * det[5 + class_id]
            label = f"{CLASS_NAMES[class_id]}: {score:.2f}"

            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 255), 2)

            (text_width, text_height), baseline = cv2.getTextSize(
                label, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 1
            )
            cv2.rectangle(
                image,
                (x1, y1 - text_height - baseline),
                (x1 + text_width, y1),
                (0, 255, 255),
                -1,
            )
            cv2.putText(
                image,
                label,
                (x1, y1 - baseline),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.5,
                (0, 0, 0),
                1,
            )

        cv2.imshow("Video", image)
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

cap.release()
cv2.destroyAllWindows()
```
